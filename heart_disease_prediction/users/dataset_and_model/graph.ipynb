```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import base64
from io import BytesIO
import joblib
```

```python
# Function to save a plot as a base64 string
def save_plot_as_base64(fig):
    buf = BytesIO()
    fig.savefig(buf, format='png')
    buf.seek(0)
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    buf.close()
    return img_base64
```

```python
# Load the dataset
data = pd.read_csv("heart.csv")
```

```python
# Dictionary to store the plots
plots = {}
```

```python
# Pairplot
sns_plot = sns.pairplot(data, hue='target', diag_kind='kde')
sns_plot.fig.suptitle('Pairplot of Heart Disease Dataset', y=1.02)
plots['pairplot'] = save_plot_as_base64(sns_plot.fig)
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Heart Disease Dataset')
plots['correlation_heatmap'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python
# Count plots for categorical features
categorical_features = ['gender', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=feature, data=data, palette='viridis')
    plt.title(f'Count plot of {feature}')
    plots[f'countplot_{feature}'] = save_plot_as_base64(plt.gcf())
    plt.close()
```

```python

# Boxplot of age vs. target
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='age', data=data, palette='coolwarm')
plt.title('Boxplot of Age vs. Target')
plots['boxplot_age_vs_target'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python
# Distribution of cholesterol
plt.figure(figsize=(10, 6))
sns.histplot(data['chol'], kde=True, color='blue', bins=20)
plt.title('Distribution of Cholesterol')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plots['histplot_cholesterol'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python
# Distribution of maximum heart rate achieved
plt.figure(figsize=(10, 6))
sns.histplot(data['thalach'], kde=True, color='green', bins=20)
plt.title('Distribution of Maximum Heart Rate Achieved')
plt.xlabel('Maximum Heart Rate Achieved')
plt.ylabel('Frequency')
plots['histplot_max_heart_rate'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python
# Scatter plot of age vs. cholesterol colored by target
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='chol', data=data, hue='target', palette='viridis')
plt.title('Scatter plot of Age vs. Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plots['scatterplot_age_vs_cholesterol'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python
# Violin plot of age distribution by target
plt.figure(figsize=(10, 6))
sns.violinplot(x='target', y='age', data=data, palette='Set2')
plt.title('Violin plot of Age Distribution by Target')
plots['violinplot_age_by_target'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python

# Bar plot of chest pain type vs. target
plt.figure(figsize=(10, 6))
sns.barplot(x='cp', y='target', data=data, palette='husl')
plt.title('Bar plot of Chest Pain Type vs. Target')
plt.xlabel('Chest Pain Type')
plt.ylabel('Target')
plots['barplot_cp_vs_target'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python
# Scatter plot of resting blood pressure vs. age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='trestbps', data=data, hue='target', palette='tab10')
plt.title('Scatter plot of Age vs. Resting Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plots['scatterplot_age_vs_trestbps'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python

# Violin plot of fasting blood sugar vs. target
plt.figure(figsize=(10, 6))
sns.violinplot(x='target', y='fbs', data=data, palette='pastel')
plt.title('Violin plot of Fasting Blood Sugar vs. Target')
plt.xlabel('Target')
plt.ylabel('Fasting Blood Sugar')
plots['violinplot_fbs_by_target'] = save_plot_as_base64(plt.gcf())
plt.close()
```

```python
# Save the plots dictionary to a joblib file
joblib.dump(plots, 'plots.joblib')
```
